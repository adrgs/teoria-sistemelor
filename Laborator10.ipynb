```python
from scipy.signal import lti, step, impulse, bode, freqresp
import matplotlib.pyplot as plt
```

```python
H = lti([1, 3], [1, 1, 1])
H.poles, H.zeros
```

```python
t, s = step(H)
plt.figure()
plt.grid(True, which='both')
plt.title('Step Response')
plt.xlabel('t[s]')
plt.ylabel('s[a]')
plt.plot(t, s)
plt.show()
```

```python
t, i = impulse(H)
plt.figure()
plt.title('Impulse')
plt.xlabel('t[s]')
plt.ylabel('impulse[a]')
plt.plot(t,i)
plt.show()
```

```python
w, m, p = bode(H)
plt.figure()
plt.title('Magnitude plot')
plt.xlabel('w[Hz]')
plt.ylabel('dB')
plt.semilogx(w,m)
plt.grid(True, which='both')
plt.show()
```

```python
w, h = freqresp(H)
plt.figure()
plt.title('Nyquist plot')
plt.xlabel('Re[s]')
plt.ylabel('Im[s]')
plt.plot(h.real, h.imag, 'b')
plt.plot(h.real, -h.imag, 'r')
plt.show()
```
