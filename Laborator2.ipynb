```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
y = np.random.normal(5.,3,100)
x = np.linspace(0,10,100)
plt.figure()
plt.plot(x,y)
plt.show()
```

```python
n, bins, res = plt.hist(y)
plt.show()
```

```python
from scipy.stats import norm
mu,std = norm.fit(bins)
print('AVG %.2f, STDEV %.2f' %(mu, std))
```

```python
l = norm.pdf(bins, mu, std)
plt.plot(bins, l, 'r--')
plt.show()
```
