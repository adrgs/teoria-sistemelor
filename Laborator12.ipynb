```python
from scipy.signal import lti, step, impulse, bode, freqresp

import matplotlib.pyplot as plt

F = lti([3],[1,4,3])

print(F.poles, F.zeros)
```

```python
t, s = step(F)
plt.figure()
plt.title('Step Response')
plt.xlabel('t[s]')
plt.ylabel('step[a]')
plt.plot(t, s)
plt.show()
```

```python
t, i = impulse(F)
plt.figure()
plt.title('Impulse')
plt.xlabel('t[s]')
plt.ylabel('impulse[a]')
plt.plot(t,i)
plt.show()
```

```python
k = len(i)-1
while i[k]==0:
    k -= 1
    
print("Amplificare %.2fs" % t[k])
```

```python
w, m, p = bode(F)
plt.figure()
plt.title('Bode magnitude')
plt.xlabel('w[Hz]')
plt.ylabel('dB')
plt.semilogx(w,m)
plt.grid(True, which='both')
plt.show()
```

```python
plt.figure()
plt.title('Phase plot')
plt.xlabel('w[Hz]')
plt.ylabel('deg')
plt.semilogx(w,p)
plt.grid(True)
plt.show()
```

```python
w, h = freqresp(F)
plt.figure()
plt.title('Nyquist Contour')
plt.xlabel('Re[s]')
plt.ylabel('Im[s]')
plt.plot(h.real, h.imag, 'b')
plt.plot(h.real, -h.imag, 'r')
plt.show()
```
