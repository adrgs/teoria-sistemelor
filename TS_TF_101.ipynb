```python
import sympy as sym
from sympy.abc import s,t,x
import numpy as np
from sympy.integrals import inverse_laplace_transform as ilt
from matplotlib import pyplot as plt
```

```python
U1 = 2/s*sym.exp(-s)
u1 = ilt(U1,s,t)
print(u1)
```

```python
tm = np.linspace(0, 8, 100)
st = np.zeros(len(tm))

for i in range(len(tm)):
    st[i] = u1.subs(t, tm[i])

plt.figure()
plt.plot(tm, st)
plt.show()
```

```python
G = 5*(s+1)/(s+3)**2
Y1 = U1*G
y1 = ilt(Y1, s,t)
s1 = np.zeros(len(tm))
for i in range(len(tm)):
    s1[i] = y1.subs(t, tm[i])
    
plt.figure()
plt.plot(tm, st, label='y1')
plt.plot(tm, s1, label='s1')
plt.legend()
plt.show()
```

```python
U2 = -2/s*sym.exp(-3*s)
u2 = ilt(U2, s, t)

for i in range(len(tm)):
    st[i] += u2.subs(t, tm[i])

plt.figure()
plt.plot(tm, st)
plt.show()
```

```python
Y2 = U2*G
y2 = ilt(Y2, s, t)

for i in range(len(tm)):
    s1[i] += y2.subs(t,tm[i])

plt.figure()
plt.plot(tm, st, label ='x(t)')
plt.plot(tm, s1, label ='y(t)')
plt.legend()
plt.show()
```

```python
U1 = 2/s*sym.exp(-s)
U2 = -1/s**2*sym.exp(-3*s)
U3 = 1/s**2*sym.exp(-5*s)

Y1 = U1*G
Y2 = U2*G
Y3 = U3*G

u1 = ilt(U1,s,t)
u2 = ilt(U2,s,t)
u3 = ilt(U3,s,t)

y1 = ilt(Y1,s,t)
y2 = ilt(Y2,s,t)
y3 = ilt(Y3,s,t)

us = np.zeros(len(tm))
for u in [u1, u2, u3]:
    for i in range(len(tm)):
        us[i] += u.subs(t, tm[i])

plt.figure()
plt.plot(tm, us)
plt.show()
```

```python
ys = np.zeros(len(tm))
for y in [y1, y2, y3]:
    for i in range(len(tm)):
        ys[i] += y.subs(t, tm[i])
        
plt.figure()
plt.plot(tm, us, label ='x(t)')
plt.plot(tm, ys, label ='y(t)')
plt.legend()
plt.show()
```
